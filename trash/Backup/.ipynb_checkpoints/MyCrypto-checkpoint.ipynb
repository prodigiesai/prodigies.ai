```python
import pandas as pd
import time
from datetime import datetime
from pytz import timezone
from binance.client import Client


def update_historical_data(symbol, interval='1h', filename='historical_data.csv'):
    # Step 1: Load existing file if it exists; otherwise, create a new one
    try:
        df_existing = pd.read_csv(filename, index_col='timestamp', parse_dates=True)

        # Apply timezone localization only if the data is not already timezone-aware
        if df_existing.index.tz is None:
            df_existing.index = df_existing.index.tz_localize('UTC').tz_convert('America/New_York')
        else:
            df_existing.index = df_existing.index.tz_convert('America/New_York')
        
        last_timestamp = df_existing.index[-1] if not df_existing.empty else pd.Timestamp("NaT")

        # Check if last_timestamp is NaT; if so, set a default start date
        if pd.isna(last_timestamp):
            last_timestamp = datetime(2024, 11, 11, tzinfo=timezone('UTC'))
            print("No existing data found; starting from default date 2024-11-11.")
        else:
            print(f"Último dato registrado: {last_timestamp}")

    except (FileNotFoundError, pd.errors.EmptyDataError):
        # If file doesn't exist or is empty, create it with a default start date
        df_existing = pd.DataFrame(columns=['timestamp', 'close', 'volume'])
        df_existing.to_csv(filename, index=False)
        last_timestamp = datetime(2024, 11, 11, tzinfo=timezone('UTC'))
        print("Archivo no encontrado o vacío. Creando un nuevo archivo de datos históricos.")

    # Step 2: Retrieve data from Binance starting from the last recorded timestamp
    new_data = []
    start_time = int(last_timestamp.timestamp() * 1000) + 1  # Start just after the last record

    while True:
        klines = client.get_klines(symbol=symbol, interval=interval, startTime=start_time, limit=1000)
        
        if not klines:
            break  # Stop if there are no more data

        # Convert retrieved data to a DataFrame
        temp_df = pd.DataFrame(klines, columns=[
            'timestamp', 'open', 'high', 'low', 'close', 'volume', 'close_time',
            'quote_asset_volume', 'number_of_trades', 'taker_buy_base_asset_volume',
            'taker_buy_quote_asset_volume', 'ignore'
        ])
        
        # Clean and transform data
        temp_df['timestamp'] = pd.to_datetime(temp_df['timestamp'], unit='ms', utc=True)
        temp_df['timestamp'] = temp_df['timestamp'].dt.tz_convert('America/New_York')
        temp_df['close'] = temp_df['close'].astype(float)
        temp_df['volume'] = temp_df['volume'].astype(float)
        temp_df.set_index('timestamp', inplace=True)
        
        # Append the required columns to new_data
        new_data.append(temp_df[['close', 'volume']])

        # Update the start time for the next API call
        start_time = int(temp_df.index[-1].timestamp() * 1000) + 1
        
        # Pause to avoid API rate limits
        time.sleep(1)

    # Step 3: Concatenate and remove duplicates if necessary
    if new_data:
        df_new = pd.concat(new_data)
        df_combined = pd.concat([df_existing, df_new]).drop_duplicates()

        # Filter data to keep only entries after the cutoff date
        cutoff_date = pd.Timestamp('2024-11-11', tz='America/New_York')
        df_combined = df_combined[df_combined.index >= cutoff_date]

        # Save the updated file
        df_combined.to_csv(filename)
        print(f"Archivo {filename} actualizado con {len(df_new)} registros nuevos.")
    else:
        print("No hay datos nuevos para añadir.")

# Call the function to update the file
update_historical_data('BTCUSDT', interval='1m', filename='historical_data.csv')

# Load the file to confirm it contains the filtered data in the correct timezone
df = pd.read_csv('historical_data.csv', index_col='timestamp', parse_dates=True)

# Check if already timezone-aware, if not, convert it
if df.index.tz is None:
    df.index = df.index.tz_localize('UTC').tz_convert('America/New_York')
else:
    df.index = df.index.tz_convert('America/New_York')

# Sort the DataFrame by the timestamp index in descending order
df = df.sort_index(ascending=False)

print(df)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import ta  # Add this line to import the 'ta' library
from ta.volatility import BollingerBands
from ta.momentum import RSIIndicator, StochasticOscillator
from ta.trend import MACD, SMAIndicator, EMAIndicator, ADXIndicator
from ta.volume import OnBalanceVolumeIndicator
from ta.trend import PSARIndicator
import matplotlib.pyplot as plt
from binance.client import Client
import time
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

# Load historical data
df = pd.read_csv('historical_data.csv', index_col='timestamp', parse_dates=True)

# Short-Term Indicators for Day Trading
bb = BollingerBands(close=df['close'], window=10, window_dev=2)
df['bb_high'] = bb.bollinger_hband()
df['bb_low'] = bb.bollinger_lband()

df['rsi'] = RSIIndicator(close=df['close'], window=5).rsi()
df['sma'] = SMAIndicator(close=df['close'], window=3).sma_indicator()
df['ema'] = EMAIndicator(close=df['close'], window=5).ema_indicator()
macd = MACD(close=df['close'], window_slow=9, window_fast=6, window_sign=3)
df['macd'] = macd.macd()
df['macd_signal'] = macd.macd_signal()
df['stoch'] = StochasticOscillator(close=df['close'], high=df['close'], low=df['close'], window=5).stoch()
df['atr'] = ta.volatility.AverageTrueRange(high=df['close'], low=df['close'], close=df['close']).average_true_range()
df['adx'] = ADXIndicator(high=df['close'], low=df['close'], close=df['close']).adx()
df['obv'] = OnBalanceVolumeIndicator(close=df['close'], volume=df['volume']).on_balance_volume()
df['psar'] = PSARIndicator(high=df['close'], low=df['close'], close=df['close']).psar()

# Signal Definitions for Buy/Sell
df['buy_signal'] = np.where((df['rsi'] < 30) & (df['close'] < df['bb_low']), 1, 0)
df['sell_signal'] = np.where((df['rsi'] > 70) & (df['close'] > df['bb_high']), -1, 0)
df['signal'] = df['buy_signal'] + df['sell_signal']

# Define Target for Next Hour Prediction
df['future_close'] = df['close'].shift(-15)  # Predicting next 15m
data = df[['rsi', 'sma', 'ema', 'macd', 'macd_signal', 'stoch', 'atr', 'adx', 'obv', 'psar', 'future_close']].dropna()

# Define features and target
features = ['rsi', 'sma', 'ema', 'macd', 'macd_signal', 'stoch', 'atr', 'adx', 'obv', 'psar']
X = data[features]
y = data['future_close']

print("Prepared data for modeling:")
print(data.head())

# Split data for training/testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Prediction
predicted_price = model.predict(X_test.iloc[-1:])
predicted_price_value = predicted_price[0]
current_price = df['close'].iloc[-1]

# Display Prediction Results
print(f"Predicted price for the next 15m: ${predicted_price_value:.2f}")
print(f"Current price: ${current_price:.2f}")

# Visualize Signals with Bollinger Bands
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['close'], label='Close Price')
plt.plot(df.index, df['bb_high'], linestyle='--', label='Bollinger High')
plt.plot(df.index, df['bb_low'], linestyle='--', label='Bollinger Low')
plt.scatter(df.index[df['buy_signal'] == 1], df['close'][df['buy_signal'] == 1], marker='^', color='g', label='Buy Signal')
plt.scatter(df.index[df['sell_signal'] == -1], df['close'][df['sell_signal'] == -1], marker='v', color='r', label='Sell Signal')
plt.legend()
plt.title('Buy/Sell Signals with Bollinger Bands and RSI')
plt.show()
```

```python
# Explanation based on prediction and signals
def trading_strategy_explanation(current_price, predicted_price, atr_value):
    if predicted_price > current_price:
        action = "BUY"
        expected_gain = predicted_price - current_price
        stop_loss = current_price - (atr_value * 1.5)  # Using 1.5x ATR for stop-loss
        reason = (
            f"The model predicts the price will increase from ${current_price:.2f} to "
            f"${predicted_price:.2f}, suggesting an opportunity to buy. "
            f"Set a stop-loss at ${stop_loss:.2f} to manage risk."
        )
    else:
        action = "SELL"
        expected_gain = current_price - predicted_price
        stop_loss = current_price + (atr_value * 1.5)  # Using 1.5x ATR for stop-loss
        reason = (
            f"The model predicts the price will decrease from ${current_price:.2f} to "
            f"${predicted_price:.2f}, suggesting an opportunity to sell. "
            f"Set a stop-loss at ${stop_loss:.2f} to manage risk."
        )

    # Print the summary explanation
    print("\n=== Trading Strategy Recommendation ===")
    print(f"Action: {action}")
    print(f"Expected Gain: ${expected_gain:.2f}")
    print(f"Recommended Stop-Loss: ${stop_loss:.2f}")
    print("Reason:", reason)

# Call the function to print the explanation
atr_value = df['atr'].iloc[-1]  # Latest ATR value for stop-loss calculation
trading_strategy_explanation(current_price, predicted_price_value, atr_value)
```
